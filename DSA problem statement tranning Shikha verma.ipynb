# 1. Check if a Number is Prime: 

  * Problem: Write a function to check if a given number is prime. 
  * Input: Number: 29 
  * Output: True 
  * Explanation: 29 is a prime number.

# Number System:-

```python
num = 29
flag = 0

for i in range(2, num):
    if num % i == 0:
        flag = 1
        break

if flag == 0:
    print("True")
else:
    print("False")
```

# OR

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Input
number = 29

# Output
print(is_prime(number))
```

# Explanation

A prime number is a number that is divisible only by 1 and itself.

The program checks if the number can be divided by any number from 2 to √n.

If divisible → not prime.

If not divisible → prime.

# Example:
29 ÷ 2, 3, 4, 5 → not divisible
So 29 is Prime.

# 2. Pythagorean Triplets: 
* Problem: Generate all Pythagorean triplets with values smaller than a given limit. 
* Input: limit = 20 
* Output: 
  * 3  4  5 
  * 8  6 10 
  * 5 12 13 
  * 15 8 17 
  * 12 16 20 

* Explanation: The triplets satisfy the condition a² + b² = c², where a, b, and c are integers. 

```python
limit = 20

for a in range(1, limit):
    for b in range(1, limit):
        for c in range(1, limit):
            if a*a + b*b == c*c:
                print(a, b, c)
```

# 6. Problem Statement 
Mayuri buys “N” no of products from a shop. The shop offers a different percentage of discount on 
each item. She wants to know the item that has the minimum discount offer, so that she can avoid 
buying that and save money.

* [Input Format: The first input refers to the no of items; the second input is the item name, price and 
discount percentage separated by comma (,) ] Assume the minimum discount offer is in the form of 
Integer. 

* Note: There can be more than one product with a minimum discount. 

# Sample Input 1: 
4 

mobile,10000,20 

shoe,5000,10 

watch,6000,15 

laptop,35000,5 

# Sample Output 1: 
shoe 

# Explanation: 
The discount on the mobile is 2000, the discount on the shoe is 500, the discount on the 
watch is 900 and the discount on the laptop is 1750. So the discount on the shoe is the minimum. 

```python
n = int(input("Enter number of items: "))

min_discount = float('inf')
product = ""

for i in range(n):
    name, price, discount = input("Enter item,price,discount: ").split(",")

    price = int(price)
    discount = int(discount)

    discount_amount = price * discount / 100

    if discount_amount < min_discount:
        min_discount = discount_amount
        product = name

print("Item with minimum discount:", product)
```

The smallest discount amount is 500.

# 7. Problem Statement – 
Bela teaches her daughter to find the factors of a given number. When she provides a number to her 
daughter, she should tell the factors of that number. Help her to do this, by writing a program. Write a 
class FindFactor.java and write the main method in it. 

# Note: 
If the input provided is negative, ignore the sign and provide the output. If the input is zero the output 
should be “No Factors”. 

# Sample Input:- 
54 
# Sample Output:- 
1,2,3,6,9,18,27,54

```python
num = int(input("Enter a number: "))

num = abs(num)

if num == 0:
    print("No Factors")
else:
    for i in range(1, num + 1):
        if num % i == 0:
            print(i, end=",")
```

# 8. Matrix Rotation by 90 Degrees: 
Rotate a 2D matrix by 90 degrees clockwise. 

Input: Matrix: [[1, 2, 3], [4, 5, 6], [7, 8, 9]] 
 
Output:Matrix:[[7, 4, 1], [8, 5, 2], [9, 6, 3]] 

Explanation: The matrix is rotated 90 degrees clockwise.

```python
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]

n = len(matrix)

# Create rotated matrix
rotated = [[0]*n for i in range(n)]

for i in range(n):
    for j in range(n):
        rotated[j][n-1-i] = matrix[i][j]

print("Rotated Matrix:")

for row in rotated:
    print(row)
```

# Program Explanation

* Step 1: Define the matrix

matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]

* Step 2: Find matrix size

n = len(matrix)

Here n = 3 because the matrix is 3 × 3.

* Step 3: Create an empty matrix

rotated = [[0]*n for i in range(n)]

This creates a new 3 × 3 matrix filled with zeros.

* Step 4: Rotate elements

rotated[j][n-1-i] = matrix[i][j]

This formula places each element in its rotated position.

Example:

matrix[0][0] = 1 → rotated[0][2]
matrix[0][1] = 2 → rotated[1][2]
matrix[0][2] = 3 → rotated[2][2]

* Step 5: Print rotated matrix

      for row in rotated:
             print(row)

             

# 9. Binary Search: 
Implement a binary search algorithm to find a target value in a sorted array.

Input: Array: [1, 2, 3, 4, 5, 6, 7, 8, 9], Target: 4 

Output: 3 

Explanation: The function returns the index of the target value in the array.

```python
arr = [1,2,3,4,5,6,7,8,9]
target = 4

low = 0
high = len(arr) - 1

while low <= high:

    mid = (low + high) // 2

    if arr[mid] == target:
        print(mid)
        break

    elif arr[mid] < target:
        low = mid + 1

    else:
        high = mid - 1
```

# 10. Problem Statement – 
You’re given an array of integers, print the number of times each integer has occurred in the array. 

# Example: 

* Input: 
10 
1 2 3 3 4 1 4 5 1 2 

* Output: 

1 occurs 3 times 

2 occurs 2 times

3 occurs 2 times 

4 occurs 2 times 

5 occurs 1 times 

```python
n = int(input("Enter number of elements: "))

arr = list(map(int, input("Enter elements: ").split()))

count = {}

for num in arr:
    if num in count:
        count[num] += 1
    else:
        count[num] = 1

for key in sorted(count):
    print(key, "occurs", count[key], "times")
```
